```python

import numpy as np,pandas as pd,matplotlib.pyplot as plt,seaborn as sns
```

```python
data=pd.read_csv("book1.csv")
data
```

```python
sns.jointplot(x="area",y="price",data=data,kind="reg")
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(data[["area"]],data.price)
```

```python
import numpy as np

# assume your scalar input data is stored in a variable called x
x = 5000

# reshape x to a 2D array with a single feature
x_2d = np.array(x).reshape(-1, 1)

x_2d
# now you can pass x_2d to the predict() method of your regression model
```

```python
y_pred = reg.predict(x_2d)
y_pred
```

need to know how it predict?

```python
value_m=reg.coef_
value_m
```

```python
value_c=reg.intercept_
value_c
```

```python
y=value_m*x_2d+value_c
```
